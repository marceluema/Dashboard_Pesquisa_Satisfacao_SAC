```python
df_sac['data'] = pd.to_datetime(df_sac['DataDaColeta'],format='%d/%m/%Y')
df_sac['ano'] = df_sac['data'].dt.year
df_sac['mes'] = df_sac['data'].dt.month
df_sac['mes_ano'] = df_sac['mes'].map(str)+ '-' + df_sac['ano'].map(str)
```

```python
df_sac['mes_ano'] = pd.to_datetime (df_sac['mes_ano'], format = '%m-%Y')
df_sac.dtypes
```

```python
media_mes_ano = df_sac.groupby (df_sac['mes_ano'], as_index=False).agg({'Satisfação_Geral': 'mean'})
media_mes_ano = media_mes_ano.round(2)
media_mes_ano
```

```python
media_aero = df_sac.groupby(df_sac['Aeroporto'], as_index=False).agg({'Satisfação_Geral': 'mean'})
media_aero = media_aero[media_aero['Satisfação_Geral']>3.5]
media_aero.sort_values ('Satisfação_Geral', ascending = False, inplace = True)
media_aero = media_aero.round(2)
media_aero
```

```python
app = dash.Dash(__name__, external_stylesheets = [dbc.themes.CYBORG])


fig_media_periodo = go.Figure (layout = {"template": "plotly_dark"})
fig_media_periodo.add_trace (go.Scatter (x =  media_mes_ano ['mes_ano'], y= media_mes_ano ['Satisfação_Geral'],
                                        name = 'Satisfação Geral (Média)'))

fig_media_periodo.update_xaxes(tickangle = 45, tickfont_size = 6 , dtick= "M1",tickformat = "%b\n%Y")

fig_media_periodo.update_yaxes (tickfont_size = 10)
fig_media_periodo.update_layout (
    title_text = 'Satisfação Geral (Média)',
    title_x = 0,
    autosize = True, 
    margin = dict(l=40,r=40,t=40,b=40)
    )

fig_rank = go.Figure (layout = {"template": "plotly_dark"}, 
                      data = [go.Table(header = dict(values= ['Aeroporto','Satisfação Geral']),
                                      cells = dict(values= [media_aero.Aeroporto, media_aero.Satisfação_Geral]))
                             ])
fig_rank.update_layout (autosize = True,
                       title_text = 'Top 10 (2017-2019)',
                       margin = dict(l=45, r=45, t=40, b=20)
                       ) 


df_corr = df_sac.iloc[:,19:74]

correlacao =  df_corr.corr()
correlacao.head(5)

correlação_satisfacao_geral = correlacao['Satisfação_Geral']
correlacao= pd.DataFrame(correlação_satisfacao_geral)
corr = correlacao.reset_index(inplace = False)
df_corr = corr[(corr['Satisfação_Geral']>0.4) & (corr['Satisfação_Geral']!=1)]
df_corr = df_corr.sort_values(['Satisfação_Geral'], ascending = True)
df_corr = df_corr.round(2) 
df_corr

fig_corr =go.Figure (layout = {"template": "plotly_dark"})
fig_corr.add_trace (go.Bar(x = df_corr['index'], y = df_corr['Satisfação_Geral'], 
                          text= df_corr['Satisfação_Geral'], textfont = dict(color= "white"), textfont_size= 10))

fig_corr.update_xaxes(tickangle = 45, tickfont_size= 8)
fig_corr.update_yaxes(tickfont_size= 8, nticks = 10, showgrid= False)

fig_corr.update_layout (
    title_text = 'Correlação (Satisfação Geral)',
    title_x =0,
    autosize = True, 
    margin = dict(l=20,r=20,t=40,b=20)
    )




app.layout = dbc.Container ([
    dbc.Row ([
        dbc.Col(html.H5 ("Pesquisa de Satisfação - Aeroportos", style ={"margin-top":'10px', 
                                                                        'margin-botton':'10px'}), md =5
               ),
        dbc.Col (md=7)
        ]),
    
    dbc.Row ([
        dbc.Col(
            dcc.Graph(id = 'fig_media_periodo', figure = fig_media_periodo, style = {"margin-top":'10px',
                                                                                     "margin-botton":"10px",'height': "40vh"})
        )      
    ]),
    
    dbc.Row ([
        dbc.Col(
            dcc.Graph(id = 'fig_rank', figure = fig_rank, style = {"margin-top":'10px',
                                                                   "margin-botton":"5px",'height': '46vh'}), md = 4
        ),
        dbc.Col(
            dcc.Graph(id ='fig_corr', figure = fig_corr, style = {"margin-top":'10px',
                                                                   "margin-botton":"10px",'height':'46vh'}), md =8
        )        
    ]),  
    
], fluid = True)

if __name__ == "__main__":
    app.run_server(debug = True, use_reloader=False)
    
```
